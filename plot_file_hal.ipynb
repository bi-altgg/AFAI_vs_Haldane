```python
import numpy as np
import matplotlib.pyplot as plt

# Define the list of text file names
file_names = [
    'LE_Hal_00.txt',
    'LE_Hal_01.txt',
    'LE_Hal_02.txt',
    'LE_Hal_03.txt',
    'LE_Hal_04.txt',
    'LE_Hal_05.txt',
    'LE_Hal_06.txt',
    'LE_Hal_07.txt',
    'LE_Hal_08.txt',
    'LE_Hal_09.txt',
    'LE_Hal_10.txt'
    # Add more filenames as needed
]

# Initialize lists to hold data and errors
data = []
errors = []

# Load each main data file and corresponding error file
for filename in file_names:
    try:
        # Load the main data
        file_data = np.loadtxt(filename)
        data.append(file_data)  # Assuming this file contains the data
    except Exception as e:
        print(f"Error loading {filename}: {e}")

    # Construct the corresponding error file name
    error_filename = filename.replace('.txt', '_err.txt')
    try:
        # Load the error data
        error_data = np.loadtxt(error_filename)
        errors.append(error_data)  # Assuming this file contains the errors
    except Exception as e:
        print(f"Error loading {error_filename}: {e}")

# Choose a base color and number of datasets
base_color = (0.5, 0.2, 0.68)  # Example base color (R, G, B)
num_datasets = len(data)

# Generate intensities
intensities = np.linspace(0.1, 1, num_datasets)

# Polynomial degree for fitting
poly_degree = 1  # You can change this as needed

# Plot each dataset with increasing intensity and error bars
for idx, (dataset, error) in enumerate(zip(data, errors)):
    x = np.arange(len(dataset))  # Create x values as indices
    # Create a color by combining base color with intensity
    color = (base_color[0] * intensities[idx], 
             base_color[1] * intensities[idx], 
             base_color[2] * intensities[idx])
    
    # Plot data with error bars
    plt.errorbar(x, dataset, yerr=error, label=f'Data {idx + 1}', 
                 color=color, fmt='-o', capsize=3)  # fmt='-o' for line with circle markers
    
    # Fit a polynomial to the data
    coeffs = np.polyfit(x, dataset, poly_degree)  # Fit polynomial

    # Print the slope and y-intercept
    slope = coeffs[0]  # For linear fit, slope is the first coefficient
    intercept = coeffs[-1]  # Y-intercept is the last coefficient
    print(f'Dataset {idx + 1}: Slope = {slope:.4f}, Y-Intercept = {intercept:.4f}')
    
    # Generate fitted values for plotting
    fitted_y = np.polyval(coeffs, x)  # Evaluate polynomial

    # Plot the fitted curve
    plt.plot(x, fitted_y, color=color, linestyle='--', linewidth=2, 
             label=f'Fit {idx + 1}')  # Dashed line for fit
    
# Customize the plot
plt.title('LE for noise realizatiobs 0 to 1 in Haldane phase')
plt.xlabel('Number of periods of evolution')
plt.ylabel('Loschmidt echo')
#plt.legend()  # Uncommenting this will enable the legend
plt.grid()
plt.show()
```
